```python
import numpy as np
import plotly.graph_objects as go
from plotly.subplots import make_subplots
data_exp = np.loadtxt('result_explicit.txt')
N_arr_exp = data_exp[:,0]
M_arr_exp = data_exp[:,1]
errors_exp= data_exp[:,2]
rate_exp=[]

for i in range(0,len(N_arr_exp)-1):
    h1 = 1.0/(M_arr_exp[i+1]-0.5)
    h0 = 1.0/(M_arr_exp[i]-0.5)
    rate_exp.append(np.log(errors_exp[i+1]/errors_exp[i])/np.log(M_arr_exp[i]/M_arr_exp[i+1]))
data_imp = np.loadtxt('result_implicit.txt')
N_arr_imp = data_imp[:,0]
M_arr_imp = data_imp[:,1]
errors_imp= data_imp[:,2]
rate_imp=[]

for i in range(0,len(N_arr_imp)-1):
    h1 = 1.0/(M_arr_imp[i+1]-0.5)
    h0 = 1.0/(M_arr_imp[i]-0.5)
    rate_imp.append(np.log(errors_imp[i+1]/errors_imp[i])/np.log(h1/h0))

print(rate_exp)
print()
print(rate_imp)
```

# Порядки сходимости для явной схемы

```python
for i in range(0,len(M_arr_exp)-1):
    print(f"Порядок сходимости для (M1,M2)=({int(M_arr_exp[i])},{int(M_arr_exp[i+1])}) есть {rate_exp[i]}")

```

# Порядки сходимости для неявной схемы

```python
for i in range(0,len(M_arr_imp)-1):
    print(f"Порядок сходимости для (M1,M2)=({int(M_arr_imp[i])},{int(M_arr_imp[i+1])}) есть {rate_imp[i]}")

```

```python
fig = make_subplots(rows=2,cols=1,subplot_titles=("Явная схема","Неявная схема"))
fig.add_trace(go.Scatter(x=N_arr_exp,y=errors_exp,mode='lines+markers',name='Явная схема'),row=1,col=1)
fig.add_trace(go.Scatter(x=N_arr_imp,y=errors_imp,mode='lines+markers',name='Неявная схема'),row=2,col=1)

fig.update_xaxes(title_text='N')
fig.update_yaxes(title_text='Error')
fig.update_layout(title_text='График зависимости ошибки реального и численного решения от N')
fig.show()
```
