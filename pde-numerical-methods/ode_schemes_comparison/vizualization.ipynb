```python
import numpy as np
import plotly.graph_objects as go
errors = np.loadtxt('result.txt')
N_arr = errors[:,0]
err1  = errors[:,1]
err2  = errors[:,2]
err3  = errors[:,3]
err4  = errors[:,4]

rate1=[]
rate2=[]
rate3=[]
rate4=[]
for i in range(0,len(N_arr)-1):
    h1 = 1.0/(N_arr[i+1])
    h0 = 1.0/(N_arr[i])
    rate1.append(np.log(err1[i]/err1[i+1])/np.log(h0/h1))
    rate2.append(np.log(err2[i]/err2[i+1])/np.log(h0/h1))
    rate3.append(np.log(err3[i]/err3[i+1])/np.log(h0/h1))
    rate4.append(np.log(err4[i]/err4[i+1])/np.log(h0/h1))
```

## Порядки сходимости для 1-ой разностной схемы

```python
for i in range(0,len(N_arr)-1):
    print(f"Порядок сходимости для ({N_arr[i]},{N_arr[i+1]}) = {rate1[i]}")

```

## Порядки сходимости для 2-ой разностной схемы

```python
for i in range(0,len(N_arr)-1):
    print(f"Порядок сходимости для ({N_arr[i]},{N_arr[i+1]}) = {rate2[i]}")

```

## Порядки сходимости для 3-ой разностной схемы

```python

for i in range(0,len(N_arr)-1):
    print(f"Порядок сходимости для ({N_arr[i]},{N_arr[i+1]}) = {rate3[i]}")

```

## Порядки сходимости для 4-ой разностной схемы

```python
for i in range(0,len(N_arr)-1):
    print(f"Порядок сходимости для ({N_arr[i]},{N_arr[i+1]}) = {rate4[i]}")

```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(x=N_arr,y=err1,mode='lines+markers',name='Схема 1'))
fig.add_trace(go.Scatter(x=N_arr,y=err2,mode='lines+markers',name='Схема 2'))
fig.add_trace(go.Scatter(x=N_arr,y=err3,mode='lines+markers',name='Схема 3'))
fig.add_trace(go.Scatter(x=N_arr,y=err4,mode='lines+markers',name='Схема 4'))

fig.update_xaxes(title_text='N')
fig.update_yaxes(title_text='Err')
fig.update_layout(title_text='График зависимости ошибки реального и численного решения диффура от N для 4 разностных схем ')
fig.show()
```
